```python
from collections import deque

def unico(nome):
    resultado = {}
    visitados = set()
    falta = deque([(nome, 0)])
    while falta:
        elemento, soma = falta.pop()
        resultado[elemento] = max(soma, resultado.get(elemento, float("-inf")))
        if elemento not in visitados and elemento in habilidade:
            for vizinho, pontos in habilidade[elemento].items():
                falta.appendleft((vizinho, pontos+soma))
        visitados.add(elemento)
        
    del resultado[nome]
    return resultado
        
```

```python
def nivel(pontos, inicial=0):
    i = inicial
    while pontos > 0:
        i += 1
        pontos -= i
    if pontos < 0:
        i -= 1
    return i
```

```python
%%unittest -p 1
assert nivel(0) == 0
assert nivel(1) == 1
assert nivel(2) == 1
assert nivel(3) == 2
assert nivel(4) == 2
assert nivel(5) == 2
assert nivel(6) == 3
assert nivel(0, 1) == 1
assert nivel(1, 1) == 1
assert nivel(2, 1) == 2
assert nivel(3, 1) == 2
assert nivel(4, 1) == 2
assert nivel(5, 1) == 3
assert nivel(0, 2) == 2
assert nivel(1, 2) == 2
assert nivel(2, 2) == 2
assert nivel(3, 2) == 3
```

```python
nome_armas_de_haste = unico("armas_de_haste")
nome_esgrima = unico("esgrima")
print("----------iguais----------")
for nome_chave in nome_armas_de_haste:
    if nome_chave in nome_esgrima:
        if nome_esgrima[nome_chave] == nome_armas_de_haste[nome_chave]:
            print( "{}: {}".format(nome_chave, nome_armas_de_haste[nome_chave]))
print("----------diferentes----------")
for nome_chave in nome_armas_de_haste:
    if nome_chave in nome_esgrima:
        if nome_esgrima[nome_chave] != nome_armas_de_haste[nome_chave]:
            print( "{}: {}, {}".format(nome_chave, nome_armas_de_haste[nome_chave], nome_esgrima[nome_chave]))
```
